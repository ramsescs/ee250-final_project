```python
import pickle

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (10,7)
import seaborn as sns

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import initializers
```

### Importing Data

```python
coins = pd.read_csv("coins.csv")
coins.head(10)
```

### Visualizing Data

```python
sns.scatterplot(x="reflectance",y="weight",data=coins, hue="denomination")
```

### Splitting Data into Train and Test Sets

```python
# test train split
X = coins[["reflectance","weight"]].values
y = coins[["denomination"]].values
from sklearn.preprocessing import LabelEncoder
yl = LabelEncoder().fit_transform(y.flatten())
X_train, X_test, y_train, y_test = train_test_split(X, yl, test_size=0.2, random_state=42)
```

### Setting up Model and ANN Layers

```python
## Import your favourite classifier and train it using the X_train data and hte y_train labels

## make sure you instantiate your classifier and name it clf

## for example, if you are using perceptron, you would do the following - 

# -> from sklearn.linear_model import Perceptron
# -> clf = Perceptron(tol=1e-3, random_state=0)
# .... call fit , etc.

# define the model
model = Sequential()
model.add(Dense(2, activation='relu', input_dim=2))
model.add(Dense(8, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

### Training Model

```python
model.fit(X_train, y_train, epochs=1000, batch_size=30, verbose=0)
```

### Test Accuracy

```python
## test the accuracy of your model
loss,acc = model.evaluate(X_test, y_test)
```

### Train Accuracy

```python
loss,acc = model.evaluate(X_train, y_train)
```

### Decision Boundary Plot

```python
## plot the decision boundary of your classifier along with the scatterplot of the training data
## hint: look at the ann example
# Plot the decision boundary
reflect = np.linspace(0,30,1000)
weight = np.linspace(3,8,100)
from itertools import product
randomdata = np.array(list(product(reflect,weight)))

g = model.predict(randomdata)

for i in range(len(g)):
    if g[i] >= 0.5:
        g[i] = 1
    else:
        g[i] = 0

db=pd.DataFrame({"reflect":randomdata[:,0],"weight":randomdata[:,1],"label": g[:,0]})
X_train,X_test,y_train,y_test = train_test_split(X,yl, test_size=0.2,random_state=42)
sns.scatterplot(data=db,x="reflect",y="weight",hue="label")
plt.scatter(X_train[:,0],X_train[:,1],color="black")
plt.xlim(20,30)
```
